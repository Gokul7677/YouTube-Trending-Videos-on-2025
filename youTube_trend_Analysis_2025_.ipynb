```python
most_liked_video = df.sort_values(by='likes', ascending=False).head(1)
print(" Most Liked Video on YouTube \n")
video = most_liked_video.iloc[0]
print(f"Title     : {video['title']}")
print(f"Channel   : {video['channel']}")
print(f"Likes     : {video['likes']}")
print(f"Views     : {video['views']}")
print(f"Comments  : {video['comments']}")
print(f"Country   : {video['country']}")
```

```python
top_channels_likes = df.groupby('channel')['likes'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
top_channels_likes.plot(kind='bar', color='lightcoral')
plt.title('Top 10 YouTube Channels by Total Likes')
plt.xlabel('Channel')
plt.ylabel('Total Likes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
top_trending_unique = df.drop_duplicates(subset='title')  # keep only one row per unique title
top_trending_unique = top_trending_unique.sort_values(by='views', ascending=False).head(10)

print("Top 10 Trending Videos on YouTube\n")
for i, video in top_trending_unique.iterrows():
    print(f"{i + 1}. Title: {video['title']}")
    print(f"   Channel: {video['channel']}   views: {video['views']}")
```

```python
import re
def remove_emojis(text):
    return re.sub(r'[^\w\s.,!?-]', '', text)

top_trending_unique['title'] = top_trending_unique['title'].apply(remove_emojis)

# Now plot
plt.figure(figsize=(12,6))
plt.barh(top_trending_unique['title'], top_trending_unique['views'], color='mediumseagreen')
plt.title('Top 10 Unique Trending YouTube Videos by Views')
plt.xlabel('Views')
plt.ylabel('Video Title')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter for India
india_df = df[df['country'] == 'IN']

# Group by channel and sum comments
top_commented_channels = india_df.groupby('channel')['comments'].sum().reset_index()

# Sort descending and get top 10
top_commented_channels = top_commented_channels.sort_values(by='comments', ascending=False).head(10)

print(top_commented_channels)

# Pie chart
plt.figure(figsize=(10,10))
plt.pie(
    top_commented_channels['comments'],
    labels=top_commented_channels['channel'],
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Top 10 Most Commented YouTube Channels in India', fontsize=14)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt


df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce')
df['publish_day'] = df['published_at'].dt.day_name()
daywise_uploads = df['publish_day'].value_counts().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

daywise_views = df.groupby('publish_day')['views'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
daywise_uploads.plot(kind='bar', color='skyblue')
plt.title('Day-wise YouTube Video Uploads')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.subplot(1, 2, 2)
(daywise_views / 1e6).plot(kind='bar', color='coral')  # Views in Millions
plt.title('Day-wise Total Views on YouTube')
plt.xlabel('Day of the Week')
plt.ylabel('Views (in Millions)')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Country-wise total metrics
country_views = df.groupby('country')['views'].sum().sort_values(ascending=False)
country_likes = df.groupby('country')['likes'].sum().sort_values(ascending=False)
country_comments = df.groupby('country')['comments'].sum().sort_values(ascending=False)

# Plot histogram
plt.figure(figsize=(16,6))

# Views Histogram
plt.subplot(1, 3, 1)
(country_views / 1e6).plot(kind='bar', color='skyblue')
plt.title('Total Views per Country')
plt.xlabel('Country')
plt.ylabel('Views (in Millions)')
plt.xticks(rotation=45)

# Likes Histogram
plt.subplot(1, 3, 2)
(country_likes / 1e6).plot(kind='bar', color='mediumseagreen')
plt.title('Total Likes per Country')
plt.xlabel('Country')
plt.ylabel('Likes (in Millions)')
plt.xticks(rotation=45)

# Comments Histogram
plt.subplot(1, 3, 3)
(country_comments / 1e6).plot(kind='bar', color='salmon')
plt.title('Total Comments per Country')
plt.xlabel('Country')
plt.ylabel('Comments (in Millions)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()



```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert fetch_date to datetime
df['fetch_date'] = pd.to_datetime(df['fetch_date'], errors='coerce')

# Select a specific day (change this if needed)
specific_day = '2025-07-03'
df_day = df[df['fetch_date'] == specific_day]

if df_day.empty:
    print(f"No data available for {specific_day}")
else:
    # Group statistics by country
    country_day_stats = df_day.groupby('country').agg({
        'views': 'sum',
        'likes': 'sum',
        'comments': 'sum',
        'video_id': 'count'
    }).rename(columns={'video_id': 'uploads'}).reset_index()

    # Plotting Pie Charts
    plt.figure(figsize=(18, 6))

    # Pie Chart for Views
    plt.subplot(1, 3, 1)
    plt.pie(country_day_stats['views'], labels=country_day_stats['country'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Blues"))
    plt.title(f'Views Share by Country on {specific_day}')

    # Pie Chart for Likes
    plt.subplot(1, 3, 2)
    plt.pie(country_day_stats['likes'], labels=country_day_stats['country'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Greens"))
    plt.title(f'Likes Share by Country on {specific_day}')

    # Pie Chart for Uploads
    plt.subplot(1, 3, 3)
    plt.pie(country_day_stats['uploads'], labels=country_day_stats['country'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Oranges"))
    plt.title(f'Uploads Share by Country on {specific_day}')

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

df['fetch_date'] = pd.to_datetime(df['fetch_date'], errors='coerce')

# Choose specific date
specific_day = '2025-07-03'
df_day = df[df['fetch_date'] == specific_day]

def remove_emojis(text):
    return re.sub(r'[^\w\s.,!?-]', '', str(text))

if df_day.empty:
    print(f"No data available for {specific_day}")
else:

    top5_views = df_day.sort_values(by='views', ascending=False).head(5).copy()
    top5_views['title'] = top5_views['title'].apply(remove_emojis)

    # Convert views to millions for X-axis
    top5_views['views_million'] = top5_views['views'] / 1_000_000

    # Plot bar chart (views in millions)
    plt.figure(figsize=(12, 6))
    sns.barplot(data=top5_views, x='views_million', y='title', color='skyblue')  # Use a fixed color

    plt.title(f'Top 5 Most Viewed Videos on {specific_day}', fontsize=14)
    plt.xlabel('Views (in Millions)')
    plt.ylabel('Video Title')
    plt.tight_layout()
    plt.show()
```
